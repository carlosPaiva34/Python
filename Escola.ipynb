```python
import matplotlib.pyplot as map
aluno=input("Nome do aluno: ")
disciplina=input("Informe a disciplina do aluno: ")
notas=[]
soma=0
#notas das ultimas três provas da disciplina 
for i in range(0,3):
    notas.append(float(input("Informe nota-"+str(i+1)+":\n")))
    soma=soma+notas[i]

media=soma/3

if media>=7.5:
    status="aprovado"
else :
    status="reprovado"
print("\n\n")
formato=f"{media:.2f}"
print("Media: "+str(formato)+"\n"+
      "Status: "+status+"\n"+
      "Aluno: "+aluno+"\n"+
      "Disciplina: "+disciplina)
map.plot(notas)
map.show()
```
